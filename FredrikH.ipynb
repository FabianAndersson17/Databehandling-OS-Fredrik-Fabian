```python
import pandas as pd
import seaborn as sns
import numpy as np
import plotly_express as px

# Reads in the csv-file and creates a dataframe.
athlete_events = pd.read_csv("Data/athlete_events.csv")
noc_regions = pd.read_csv("Data/noc_regions.csv")
athlete_events.head()
```

```python
#Uses "len" to get the amount of countries in the list.
print(f"Exercise 0a: There are {len(noc_regions)} countries in the dataset.")

#Using a for-loop to go through the list of countries and print them.
print("Exercise 0b:")
for i in range(0,len(noc_regions)):
    print(noc_regions.iloc[i, 0],end=" ")

```

```python
fig = px.pie(athlete_events,names='Sex', title='Exercise 0g :Split Men / Female of all competitors for all competitions since start')
fig.show()
```

```python
athlete_events_drop = athlete_events.drop_duplicates(subset=["Event", "Games", "Medal"])
athlete_events_grouped = athlete_events_drop.groupby("NOC").count() #Grouping all columns based on "NOC" (shortcut for Countries).
athlete_events_grouped = athlete_events_grouped.sort_values(by="Medal", ascending = False)
athlete_events_grouped_top10=  athlete_events_grouped.reset_index().head(10)
fig1 = px.bar(athlete_events_grouped_top10,x= "NOC", y = "Medal", title = "Exercise 0h: Top 10 countries with most medals in Olympic Games")
fig1.show()
```

```python
athlete_events_all = athlete_events.groupby("NOC").count() #Grouping all columns based on "NOC" (shortcut for Countries).
athlete_events_all = athlete_events_all.sort_values(by="Medal", ascending = False)
athlete_events_all=  athlete_events_all.reset_index().head(10)
athlete_events_all=  athlete_events_all.rename({"Medal" : "Participants"}, axis = "columns")
fig2 = px.bar(athlete_events_all,x= "NOC", y = "Participants", title = "Exercise 0i: Top 10 countries with most participants in Olympic Games")
fig2.show()
```

```python
athlete_events_per_year = athlete_events_drop.groupby(["NOC","Year"]).count()
athlete_events_per_year =  athlete_events_per_year.reset_index()
russia_sovjet_merge = athlete_events_per_year[athlete_events_per_year["NOC"].isin(["RUS", "URS"])].reset_index()
russia_sovjet_merge = russia_sovjet_merge.sort_values("Year")
germany_medals_per_year = athlete_events_per_year[athlete_events_per_year["NOC"].isin(["RUS"])]
usa_medals_per_year = athlete_events_per_year[athlete_events_per_year["NOC"].isin(["USA"])]
china_medals_per_year = athlete_events_per_year[athlete_events_per_year["NOC"].isin(["CHN"])]

#Plotting the sorted datasets
fig3 = px.line(title = "Exercise 0i: Medals per year per choosen countries")
rus_sov_line = fig3.add_scatter(y = russia_sovjet_merge["Medal"], x = russia_sovjet_merge["Year"])
usa_line = fig3.add_scatter(y = usa_medals_per_year["Medal"], x = usa_medals_per_year["Year"])
china_line = fig3.add_scatter(y = china_medals_per_year["Medal"], x = china_medals_per_year["Year"])
fig3.show()
```
